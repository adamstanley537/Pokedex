```python
poke_list = ["Snorlax"]
```

```python
for poke in poke_list:
    
    file_list = os.listdir('.' + "/" + poke)

    duplicates = []
    hash_keys = dict()
    for index, filename in  enumerate(file_list):
        if os.path.isfile('.' + "/" + poke + "/" + filename):

            with open('.' + "/" + poke + "/" + filename, 'rb') as f:
                filehash = hashlib.md5(f.read()).hexdigest()

            if filehash not in hash_keys: 
                hash_keys[filehash] = index

            else:
                duplicates.append((index,hash_keys[filehash]))
                
#     print(poke, len(duplicates))
    for index in duplicates:
        print('.' + "/" + poke + "/" + file_list[index[0]])
        os.remove('.' + "/" + poke + "/" + file_list[index[0]])
```

```python
batch_size = 32
img_height = 224
img_width = 224
```

```python
train_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.1,
    subset="training",
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size,
    color_mode="rgb",
    label_mode="categorical"
)
```

```python
val_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.1,
    subset="validation",
    seed=123,
    image_size=(img_height, img_height),
    batch_size=batch_size,
    color_mode="rgb",
    label_mode="categorical")
```
